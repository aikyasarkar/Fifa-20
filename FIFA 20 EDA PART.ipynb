## Analysis of Players salary

```python
player_salary = data[['short_name','wage_eur']]
player_salary.head()
```

```python
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(player_salary['short_name'][0:5],player_salary['wage_eur'][0:5])
plt.xticks(rotation=30)
plt.xlabel('Player')
plt.ylabel('wage_eur');
```

```python
#shooting

player_shot = data[['short_name','shooting']]
player_shot.sort_values(by=['shooting'],ascending=False).head()
```

```python
#defending

player_defend = data[['short_name','nationality','club','defending']]
player_defend.sort_values(by=['defending'],ascending=False).head()
```

```python
#passing

player_pass = data[['short_name','nationality','club','passing']]
player_pass.sort_values(by=['passing'],ascending=False).head()
```

```python
# youngest palyers
player_age = data[['short_name','age']]
player_age.sort_values(by=['age']).head()
```

```python
#best youngest players
data_young = data[data['age']<=20].sort_values(by=['potential','value_eur'],ascending=[False,True])[0:5]
plt.figure(figsize=(6,6))
sns.barplot(y='short_name',x='wage_eur',data=data_young)
```
